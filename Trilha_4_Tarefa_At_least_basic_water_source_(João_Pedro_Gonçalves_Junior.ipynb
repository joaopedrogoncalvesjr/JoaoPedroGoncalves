Dataframe: **At least basic water source, urban access (%)**, 21 randomly selected countries.

The percentage of people using at least basic water services. This indicator encompasses both people using basic water services as well as those using safely water services. Basic drinking water services is defined as drinking water from an imporved source, provided collection time is not more than 30 minutes for a round trip. Improved water sources include piped water, boreholes or tubewells, protected dug wells, proteted springs and packaged or delivered water.

1. Which country presented the lowest urban population with access to water in 2014? Answer: Angola, with only 70.2% of the population having access to water.

2. Considering the 21 observed countries in 2014, most of them are in which range o % representing population access to water? Answer: Most of the countries in the dataset in 2014 presented a percentual that varies between 90% and 95% of total population with access to water. That became visible in the Scatterplot.

3. How many countries with less than 80% of population with water access?
Answer: Only 2 (Among the total 21 in the dataset). That can be seen in the histogram. 

4. What is the biggest outliar from the dataset? Answer: Also visible in the Scatterplot but even more evident in the boxplot: 70%, coming from the Angola in 2014.

## Plotagem de gráficos

Gráfico de dispersão com as variáveis **wt** e **mpg**

```python
plt.figure(figsize=(15,8))
plt.scatter(df['2014'], df['country']);
```

Adicionando **Título**, **Suptítulo** e **rótulos dos eixos**

```python
plt.figure(figsize=(15,8))
plt.scatter(df['2014'], df['country']);
plt.title("Data from 2014, Gapminder Website")
plt.suptitle("At least basic water source, urban access")
plt.xlabel("%")
plt.ylabel( "Country")
```

Histograma de uma variável, para isso, vamos usar a variável **2014** (ano)

```python
plt.figure(figsize=(15,8))
plt.hist(df['2014'], edgecolor="black");

plt.title("Volume of countries with water access histogram", fontsize=20)
plt.xlabel("% of urban population with water access", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
```

Histograma com os dados normalizados, e mais uma vez, utilizaremos a variável **2014**

```python
plt.figure(figsize=(15,8))
plt.hist(df['2014'], edgecolor="black", density=True)
```

```python
plt.figure(figsize=(15,8))
plt.hist(df['2014'], edgecolor="black", orientation="horizontal")
```

Duas linhas verticais, a partir de valores calculados. São cálculos de Média e Mediana, a partir da variável **2014**

```python
plt.figure(figsize=(15,8))

cores = ["orange","purple"]
legenda = ["Média","Mediana"]

plt.hist(df['2014'], edgecolor="black");
media, mediana = df['2014'].mean(), df['2014'].median()
plt.axvline(media, color=cores[0], lw=3, ls="--")
plt.axvline(mediana, color=cores[1], lw=3, ls="--")
```

Adicionando legenda no gráfico

```python
plt.figure(figsize=(15,8))

cores = ["orange","purple"]
legenda = ["Média","Mediana"]

plt.hist(df['2014'], edgecolor="black");
media, mediana = df['2014'].mean(), df['2014'].median()
plt.axvline(media, color=cores[0], lw=3, ls="--")
plt.axvline(mediana, color=cores[1], lw=3, ls="--")

plt.legend(legenda, loc="upper right")
```

```python
plt.figure(figsize=(15,8))

cores = ["orange","purple"]
legenda = ["Média","Mediana"]

plt.hist(df['2014'], edgecolor="black");
media, mediana = df['2014'].mean(), df['2014'].median()
plt.axvline(media, color=cores[0], lw=3, ls="--")
plt.axvline(mediana, color=cores[1], lw=3, ls="--")

plt.legend(legenda, loc="upper right")

plt.annotate(legenda[0], xy=(media + 0.5, 6),
             color=cores[0], fontsize=10)
plt.annotate(legenda[1], xy=(mediana - 3,6),
             color=cores[1], fontsize=10)
```

Gráfico de linha, com as variáveis **wt** e **mpg**

```python
plt.figure(figsize=(15,8))

plt.plot(df['2014'], df['country'], color="green", ls = "dotted")
plt.title("At least basic water source", fontsize=25)
plt.xlabel("%", fontsize=15)
plt.ylabel("Countries", fontsize=15)
plt.legend(["population %"], loc="upper right")
```
